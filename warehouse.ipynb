```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Desktop/warehouse.csv')

df
```

```python
df.info()
```

```python
# converting datetiem feature into pandas datetime object
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
# Find out closng quantity of each Category for Given date

# quantities in KG
# Id is unique 
```

```python
# all categories
all_material = df['material_name'].unique()

all_material
```

```python
# input date
given_date = '2023-11-04' 

# filtered data with the given date
date_filtered_data = df[df['datetime'].dt.date == pd.Timestamp(given_date)]   


# looping to find last remaining closing quantity for each category
closing_quantities = {}
for material in all_material:
    material_data = date_filtered_data[date_filtered_data['material_name'] == material]
    
    
    # if category is not presented at given date then we use last_transaction of that category before given date
    if len(material_data) == 0:
        temp_filtered = df[(df['datetime'].dt.date <= pd.Timestamp(given_date)) & (df['material_name'] == material)] 
        last_transaction = temp_filtered.sort_values(by='datetime').iloc[-1]
    
    # if already present then getting its last transcation of given date
    else:
        last_transaction = material_data.sort_values(by='datetime').iloc[-1]
      
    # then getting the closing quantity from the last transaction
    closing_qty = last_transaction['closing_qty']
    # appending closing quantity to dictionary
    closing_quantities[material] = closing_qty
    
    
# These are the categories with their closing quantity on the given date
closing_quantities
```
